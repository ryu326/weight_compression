```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#      (   )
# from matplotlib import font_manager, rc
# font_path = "C:/Windows/Fonts/malgun.ttf"  #  OS    
# font = font_manager.FontProperties(fname=font_path).get_name()
# rc('font', family=font)
# plt.rcParams['axes.unicode_minus'] = False #    


# 1.     (: 1000x800 )
#      0,     
np.random.seed(42)
matrix = np.random.randn(1024, 4096)

# 2. (row) (column)  (std) 
# axis=1:  , axis=0:  
row_std_distribution = matrix.std(axis=1)
col_std_distribution = matrix.std(axis=0)

# 3.  
# 1 2 (subplot) 
fig, axes = plt.subplots(2, 1, figsize=(6, 10))
# fig.suptitle('   ', fontsize=16) #   

# : (Row)   
sns.histplot(ax=axes[0], data=row_std_distribution, kde=True)
axes[0].set_title('Row Standard Deviation Distribution')
axes[0].set_xlabel('Standard Deviation Value')
axes[0].set_ylabel('Frequency')

# : (Column)   
sns.histplot(ax=axes[1], data=col_std_distribution, kde=True)
axes[1].set_title('Column Standard Deviation Distribution')
axes[1].set_xlabel('Standard Deviation Value')
axes[1].set_ylabel('Count')

#     
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # suptitle   
plt.show()
```
