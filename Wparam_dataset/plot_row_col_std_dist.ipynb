```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 한국어 폰트가 깨지지 않도록 설정 (필요 시 주석 해제)
# from matplotlib import font_manager, rc
# font_path = "C:/Windows/Fonts/malgun.ttf"  # 사용하는 OS에 맞는 폰트 경로로 수정
# font = font_manager.FontProperties(fname=font_path).get_name()
# rc('font', family=font)
# plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지


# 1. 분석할 샘플 행렬 생성 (예: 1000x800 크기)
# 이미지와 유사한 분포를 위해 평균 0, 작은 표준편차를 갖는 정규분포로 생성
np.random.seed(42)
matrix = np.random.randn(1024, 4096)

# 2. 행(row)과 열(column) 별로 표준편차(std) 계산
# axis=1: 행별 계산, axis=0: 열별 계산
row_std_distribution = matrix.std(axis=1)
col_std_distribution = matrix.std(axis=0)

# 3. 분포 시각화
# 1행 2열의 서브플롯(subplot) 생성
fig, axes = plt.subplots(2, 1, figsize=(6, 10))
# fig.suptitle('행과 열의 표준편차 분포', fontsize=16) # 전체 플롯 제목

# 왼쪽: 행(Row) 표준편차 분포 플롯
sns.histplot(ax=axes[0], data=row_std_distribution, kde=True)
axes[0].set_title('Row Standard Deviation Distribution')
axes[0].set_xlabel('Standard Deviation Value')
axes[0].set_ylabel('Frequency')

# 오른쪽: 열(Column) 표준편차 분포 플롯
sns.histplot(ax=axes[1], data=col_std_distribution, kde=True)
axes[1].set_title('Column Standard Deviation Distribution')
axes[1].set_xlabel('Standard Deviation Value')
axes[1].set_ylabel('Count')

# 레이아웃을 깔끔하게 조정하고 플롯 출력
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # suptitle과 겹치지 않도록 조정
plt.show()
```
